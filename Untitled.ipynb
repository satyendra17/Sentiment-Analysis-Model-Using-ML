```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import preprocess_functions as ppf
```

```python
df = pd.read_csv('https://github.com/laxmimerit/All-CSV-ML-Data-Files-Download/raw/master/IMDB-Dataset.csv')
```

```python
df.head()
```

```python
df['word_counts'] = df['review'].apply(lambda x: ppf.get_wordcounts(x))
df['char_counts'] = df['review'].apply(lambda x: ppf.get_charcounts(x))
df['avg_wordlength'] = df['review'].apply(lambda x: ppf.get_avg_wordlength(x))
df['stopwords_counts'] = df['review'].apply(lambda x: ppf.get_stopwords_counts(x))

df['review'] = df['review'].apply(lambda x: ppf.remove_html_tags(x))
df['review'] = df['review'].str.lower()
```

```python
df['review_without_stopwords'] = df['review'].apply(lambda x: ppf.remove_stopwords(x))
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
df.columns
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(hue = 'sentiment', x= 'word_counts', data = df, kde = True)
plt.title('Word counts')



plt.subplot(2,2,2)
sns.histplot(hue = 'sentiment', x= 'char_counts', data = df, kde = True)
plt.title('char counts')


plt.subplot(2,2,3)
sns.histplot(hue = 'sentiment', x= 'avg_wordlength', data = df, kde = True)
plt.title('Average Word Length')


plt.subplot(2,2,4)
sns.histplot(hue = 'sentiment', x= 'stopwords_counts', data = df, kde = True)
plt.title('Stopwords counts')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report , confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
import warnings
from sklearn.svm import SVC
warnings.filterwarnings('ignore')
```

```python
X_train , X_test , y_train, y_test = train_test_split(df['review'], df['sentiment'], test_size=0.2, random_state = 0, stratify = df['sentiment'])
```

```python
X_train.shape , X_test.shape
```

```python
clf = Pipeline([('tfidf', TfidfVectorizer()), ('clf', LinearSVC())])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
svc_acc = accuracy_score(y_pred , y_test)
print("SVC accuracy: {:.2f}%".format(svc_acc*100))
```

```python
print(classification_report(y_test , y_pred))
```

```python
from mlxtend.plotting import plot_confusion_matrix

cm = confusion_matrix(y_test , y_pred)
plot_confusion_matrix(cm, class_names= clf.classes_)
```

```python
# now seeing other models as well like Logistic Regression , Naive Bayes etc

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
```

```python
# Text vectorization using TF-IDF
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

```

```python
# Logistic Regression

logreg = LogisticRegression()
logreg.fit(X_train_vectorized, y_train)

# Make predictions
logreg_pred = logreg.predict(X_test_vectorized)

# Calculate accuracy
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Logistic Regression Accuracy::{:.2f}%".format( logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print( classification_report(y_test, logreg_pred))
```

```python
from mlxtend.plotting import plot_confusion_matrix

cm = confusion_matrix(y_test , logreg_pred)
plot_confusion_matrix(cm, class_names= clf.classes_)
```

```python
# Multinomial Naive Bayes

mnb = MultinomialNB()
mnb.fit(X_train_vectorized, y_train)

# Make predictions
mnb_pred = mnb.predict(X_test_vectorized)

# Calculate accuracy
mnb_acc = accuracy_score(mnb_pred, y_test)
print("Multinomial NB Accuracy::{:.2f}%".format( mnb_acc*100))
```

```python
print(confusion_matrix(y_test, mnb_pred))
print("\n")
print( classification_report(y_test, mnb_pred))
```

```python
from mlxtend.plotting import plot_confusion_matrix

cm = confusion_matrix(y_test , mnb_pred)
plot_confusion_matrix(cm, class_names= clf.classes_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# RandomForest
random_forest = RandomForestClassifier()
random_forest.fit(X_train_vectorized, y_train)

# Make predictions
random_forest_pred = random_forest.predict(X_test_vectorized)

# Calculate accuracy
random_forest_acc = accuracy_score(random_forest_pred, y_test)
print("Random Forest Accuracy: {:.2f}%".format(random_forest_acc * 100))
```

```python
print(confusion_matrix(y_test, random_forest_pred))
print("\n")
print( classification_report(y_test, random_forest_pred))
```

```python
cm = confusion_matrix(y_test , random_forest_pred)
plot_confusion_matrix(cm, class_names= clf.classes_)
```
